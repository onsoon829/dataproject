where

```python
a = np.array([10,20,30,40,50])
a
```

```python
#where의 인자는 조건이다.
# 조건만 넣을 경우 True의 위치를 리턴한다.
np.where(a>30)
```

```python
# where의 인자 개수에 따라서 기능이 달라진다.
# 조건이 True이면 0, False이면 100을 리턴한다
np.where(a>30, 0, 100) # np.where(조건, 참일 경우의 값, 거짓일 경우의 값)
```
